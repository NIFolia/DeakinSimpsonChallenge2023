# Rename file directory

```python
import os
```

```python
def count_files(directory):
    if not os.path.isdir(directory):
        print("Invalid directory path.")
        return

    files = os.listdir(directory)
    file_count = len(files)
    print(f"There are {file_count} files in the directory.")
```

```python
def rename_images(directory):
    if not os.path.isdir(directory):
        print("Invalid directory path.")
        return

    # Get a list of all files in the directory
    files = os.listdir(directory)
    image_files = [file for file in files if file.lower().endswith(('.jpg', '.jpeg', '.png', '.gif'))]

    # Sort the image files alphabetically
    image_files.sort()

    # Rename the image files with a sequence number
    for i, file in enumerate(image_files, start=1):
        extension = os.path.splitext(file)[1]
        new_name = str(i).zfill(4) + extension  # Use zfill to pad the number with zeros
        old_path = os.path.join(directory, file)
        new_path = os.path.join(directory, new_name)
        os.rename(old_path, new_path)
        print(f"Renamed {file} to {new_name}")
```

```python
# Call the function to count the files
count_files(dire)
```
